```python
import polars as pl
```

```python
clusters = pl.read_parquet("./features_ds/ETHUSDT_features_136_23032023.parquet").sort("time")
clusters.head()
```

```python
klines = pl.read_parquet("./features_ds/ETHUSDT_KLINES_136_23032023.parquet").sort("time")
klines = klines.with_columns(pl.col("time").dt.truncate("1m"))
klines = klines.with_columns(pl.col("time") + pl.duration(minutes=1))
klines.head()
```

```python
index_price = pl.read_parquet("./features_ds/ETHUSDT_INDEX_PRICE_136_23032023.parquet").sort("time").rename({"open": "index_price_open", "high": "index_price_high", "low": "index_price_low", "close": "index_price_close", })
index_price = index_price.with_columns(pl.col("time").dt.truncate("1m"))
index_price = index_price.with_columns(pl.col("time") + pl.duration(minutes=1))
```

```python
mark_price = pl.read_parquet("./features_ds/ETHUSDT_MARK_PRICE_136_23032023.parquet").sort("time").rename({"open": "mark_price_open", "high": "mark_price_high", "low": "mark_price_low", "close": "mark_price_close", })
mark_price = mark_price.with_columns(pl.col("time").dt.truncate("1m"))
mark_price = mark_price.with_columns(pl.col("time") + pl.duration(minutes=1))
```

```python
premium_index = pl.read_parquet("./features_ds/ETHUSDT_PREMIUM_INDEX_136_23032023.parquet").sort("time").rename({"open": "premium_index_open", "high": "premium_index_high", "low": "premium_index_low", "close": "premium_index_close", })
premium_index = premium_index.with_columns(pl.col("time").dt.truncate("1m"))
premium_index = premium_index.with_columns(pl.col("time") + pl.duration(minutes=1))
```

```python
dataset = clusters.join(klines, on="time")
dataset = dataset.join(index_price, on="time")
dataset = dataset.join(mark_price, on="time")
dataset = dataset.join(premium_index, on="time")
```
